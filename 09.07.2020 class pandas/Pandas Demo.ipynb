```python
import pandas as pd
```

```python
df=pd.read_csv("employee_details.csv")
df
```

```python
#Create full name column with by concatenating first_name and last_name
df["full_name"]=df["first_name"]+" "+df["last_name"]
df
```

```python
#Create annual salary column
df["annual_salary"]=df["monthly_sal"]*12
df
```

```python
#Create gender2 column and store m if male else f

#def getGender(gen):
#    if(gen=="Male"):
#        return "M"
#    else:
#        return "F"

#apply function is used to apply your user defined function on dataframe column
df["gender2"]=df["gender"].apply(lambda gen:"M" if gen=="Male" else "F")
df
```

```python
df.to_csv("employee_details2.csv",index=False)
```
