```python
week1=["mon","tue","wed","thu","fri","sat","sun"]
temp1=[28,27,28,28,27,29,30]
```

```python
week2=["mon","tue","wed","thu","fri","sat","sun"]
temp2=[27,26,25,26,27,26,27]
```

```python
import matplotlib.pyplot as plt 
```

```python
plt.plot(week1,temp1,color="red",label="Week1")
plt.plot(week2,temp2,color="blue",label="Week2")
plt.xlabel("Week")
plt.ylabel("Temp")
plt.legend()
plt.title("Week1 VS Week2")
```
